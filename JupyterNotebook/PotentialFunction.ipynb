```python
import matplotlib.pyplot as plt
from sklearn import neighbors, datasets
import numpy as np
import numpy.random as rnd
from matplotlib.colors import ListedColormap
get_ipython().run_line_magic('matplotlib', 'inline')
```

```python
def potential_function(X, y, K, h, t, u):
    """Classifies as maximum of summary potential given by around sample objects

    Keyword arguments:
    X -- training sample
    y -- classes of training samples
    K -- Kernel function
    h -- array of each sample object window radius
    t -- array of each sample object potential multyplier
    u -- object to classify
    """
    p = []
    for i in range(0, len(X)):
        sumt = 0
        for l in range(0,len(X[i])):
            sumt += (X[i][l] - u[l])**2
            
        sumt = np.sqrt(sumt)    
        p.append(t[i]*K(sumt / h[i], 1))
    bc = np.bincount(y, p)
    res = np.argmax(bc)
    if(bc[res]==0):
        return -1
    return res


def learn_for_potential_function(X, y, p, K):
    """Finds optimal radius and potential arrays

    Keyword arguments:
    X -- training sample
    y -- classes of training samples
    p -- maximum percentage of error object we need to approach.
    """
    diff = 1
    u = [[], 0]
    t = []
    t = [0 for x in X]
    h = [0.8 for x in X]
    stopsignal = -1
    while(diff>p and stopsignal!=0):
        cd = 0
        for i in range(0, len(X)):
            u[0] = X[i]
            u[1] = y[i]
            if (potential_function(X, y, K, h, t, u[0]) != u[1]):
                cd = cd + 1
                t[i] = t[i]+1
        diff = cd/len(X)
        print(str(diff*100)+"% :"+str(cd)+" errs of "+str(len(X)))
        stopsignal -= 1
    return h,t

def k(x,l):
    if(x>l):
        return 0;
    else:
        return (l-x)/l;
```

```python
iris = datasets.load_iris()
X = iris.data[:,:2]
y = iris.target

XFull = iris.data[:,:2]
yFull = iris.target

hs,t = learn_for_potential_function(XFull,yFull,0.07,k)
```

```python
h = 0.04

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1


xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = np.array([potential_function(X, y, k, hs, t, u) for u in  np.c_[xx.ravel(), yy.ravel()]])
Z = Z.reshape(xx.shape)
```

```python
potMax = t[np.argmax(t)]
sizes = [100*v/potMax for v in t]
plt.figure(figsize=(15,15))
plt.pcolormesh(xx, yy, Z, cmap=ListedColormap(['#777777','#FFAAAA', '#AAFFAA', '#AAAAFF']))
plt.scatter(X[:,0],X[:,1], c=y, cmap=ListedColormap(['#FF0000', '#00FF00', '#0000FF']))#, s = sizes)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Potential function")
plt.show()
```
